```python
import numpy as np
import torch
import torch.nn as nn

print(torch.__version__)
```

```python
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=np.float32)
ys = np.array([-3, -1, 1, 3, 5, 7], dtype=np.float32)

X = torch.tensor(xs).reshape(-1, 1)
Y = torch.tensor(ys).reshape(-1, 1)

print(f"X: {X}")
print(f"Y: {Y}")
```

```python
model = nn.Linear(1, 1)
print(model.weight)
print(model.bias)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
loss_fn = nn.MSELoss()

for epoch in range(500):
    optimizer.zero_grad()
    predictions = model(X)
    loss = loss_fn(predictions, Y)
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.6f}")
```

```python
print(f"Weight: {model.weight.item():.4f}")
print(f"Bias: {model.bias.item():.4f}")

test_input = torch.tensor([[10.0]])
prediction = model(test_input)
print(f"Prediction for x=10: {prediction.item():.4f}")
```
